funciones en la forma list(map(estandariza,nombre de la lista)), donde estandariza es la función

que estandariza datos, y "nombre de la lista" es el nombre de una lista en particular, cuyos

datos deseamos estandarizar

y desviación estándar $\sigma$ de la lista nplista1

Fórmula para cuantiles:

$Q_i = L_{i-1}+\frac{\frac{iN}{Z}-N_{i-1}}{n_i}C_i$, donde $N$ es el número total de datos, $1\leq i\leq Z-1$,

$Z$ es el número total de subconjuntos de la partición de los $N$ datos, $L_i$ es el límite inferior de la $i$-ésima

clase, $N_i$ es la $i$-ésima frecuencia acumulada, $n_i$ es la frecuencia de la $i$-ésima clase, y $C_i$ es en ancho de la

$i$-ésima clase.


```python
mediana = pd_aparcamiento["Tiempo de estacionamiento"][1][0]+((3075.0-acumulacion[0])/np_NumVehiculos[1])*(pd_aparcamiento["Tiempo de estacionamiento"][1][1]-pd_aparcamiento["Tiempo de estacionamiento"][1][0])
```
