Una distribuci√≥n de probabilidad es una funci√≥n $p:Œ©‚Üí[0,1]$, donde $Œ©$ es el espacio muestral del fen√≥meno aleatorio. Esto es, el conjunto de resultados individuales posibles (tambi√©n llamados $\textit{puntos muestrales}$)

Ensayos de Bernoulli. Fen√≥meno aleatorio que admite dos resultados posibles con probabilidades complementarias, $p$ y $q=1-p$, que suman $1$. Los ensayos de Bernoulli tambi√©n identifican distribuciones binomiales. El siguiente c√≥digo genera una distribuci√≥n binomial mediante la librer√≠a numpy.

```python
# prompt: ensayos de Bernoulli

import numpy as np

def bernoulli_trials(n, p):
  """
  Simulates n Bernoulli trials with probability of success p.

  Args:
      n: The number of trials.
      p: The probability of success in each trial.

  Returns:
      A NumPy array of length n representing the outcomes of the trials
      (1 for success, 0 for failure).
  """
  return np.random.binomial(1, p, n)

# Example usage:
n_trials = 10  # Number of trials
probability_of_success = 0.6  # Probability of success in each trial
results = bernoulli_trials(n_trials, probability_of_success)
print(results)

# Calculate the number of successes
num_successes = np.sum(results)
print("Number of successes:", num_successes)

# Calculate the frequency of success
frequency = num_successes / n_trials
print("Frequency of success:", frequency)


```

Analicemos algunas funciones relacionadas con la distribuci√≥n binomial, como son $\textit{probability mass function}\, (\textit{pmf})$, $\textit{cumulative distribution function}\, (cdf)$, $\textit{random variable sample}\, (rvs)$, $\textit{mean}$ y $\textit{standard deviation}\, (std)$. Para hacer uso de estas funciones, necesitamos generar una distribuci√≥n binomial, a trav√©s de la librer√≠a scipy.stats

```python
# prompt: distribuci√≥n binomial

import numpy as np
from scipy.stats import binom

# Par√°metros de la distribuci√≥n binomial
n = 10  # N√∫mero de ensayos
p = 0.5  # Probabilidad de √©xito en cada ensayo

# Calcular la probabilidad de k √©xitos
k = 5  # N√∫mero de √©xitos deseados
probabilidad = binom.pmf(k, n, p)
print(f"La probabilidad de obtener {k} √©xitos en {n} ensayos es: {probabilidad}")


# Calcular la probabilidad acumulativa hasta k √©xitos
probabilidad_acumulada = binom.cdf(k, n, p)
print(f"La probabilidad de obtener hasta {k} √©xitos en {n} ensayos es: {probabilidad_acumulada}")


# Generar una muestra aleatoria de la distribuci√≥n binomial
muestra = binom.rvs(n, p, size=100)
print(f"Una muestra aleatoria de 100 valores: {muestra}")

# Calcular la media y la desviaci√≥n est√°ndar de la distribuci√≥n
media = binom.mean(n, p)
desviacion_estandar = binom.std(n, p)

print(f"Media de la distribuci√≥n: {media}")
print(f"Desviaci√≥n est√°ndar: {desviacion_estandar}")
```

```python
# prompt: gr√°fica de una distribuci√≥n binomial

import numpy as np
import matplotlib.pyplot as plt

# Valores de k (n√∫mero de √©xitos)
k_values = np.arange(0, n + 1)

# Probabilidades para cada valor de k
probabilidades = binom.pmf(k_values, n, p)

# Crear la gr√°fica de barras
plt.plot(k_values, probabilidades)

# Etiquetas y t√≠tulo
plt.xlabel("N√∫mero de √©xitos (k)")
plt.ylabel("Probabilidad")
plt.title("Distribuci√≥n Binomial (n={}, p={})".format(n, p))

# Mostrar la gr√°fica
plt.show()
```

Distribuci√≥n de Poisson. La distribuci√≥n de Poisson es una distribuci√≥n de probabilidad discreta, cuyo dominio son los n√∫meros naturales. Esta distribuci√≥n tiene el par√°metro $\lambda$, que es el valor promedio de eventos de un experimento aleatorio restringido a una temporalidad fija (o a alguna otra variable de dimensiones fijas). Matem√°ticamente, tenemos la siguiente funci√≥n $f(\lambda,k) = \frac{\lambda^ke^{-\lambda}}{k!}$. En esta funci√≥n, $k$ es un par√°metro. El valor $f(\lambda,k)$ se interpreta como la probabilidad de observar $k$ eventos dentro de la temporalidad fija (o variable de dimensiones fijas).

```python
# prompt: distribuci√≥n de Poisson

from scipy.stats import poisson
import matplotlib.pyplot as plt
import numpy as np

# Par√°metros de la distribuci√≥n de Poisson
lambda_param = 3  # Valor promedio de eventos en un intervalo dado

# Calcular la probabilidad de k eventos
k = 2  # N√∫mero de eventos deseados
probabilidad = poisson.pmf(k, lambda_param)
print(f"La probabilidad de obtener {k} eventos es: {probabilidad}")

# Calcular la probabilidad acumulativa hasta k eventos
probabilidad_acumulada = poisson.cdf(k, lambda_param)
print(f"La probabilidad de obtener hasta {k} eventos es: {probabilidad_acumulada}")

# Generar una muestra aleatoria de la distribuci√≥n de Poisson
muestra = poisson.rvs(lambda_param, size=100)
print(f"Una muestra aleatoria de 100 valores: {muestra}")

# Calcular la media y la desviaci√≥n est√°ndar de la distribuci√≥n
media = poisson.mean(lambda_param)
desviacion_estandar = poisson.std(lambda_param)
print(f"Media de la distribuci√≥n: {media}")
print(f"Desviaci√≥n est√°ndar: {desviacion_estandar}")

# Valores de k (n√∫mero de eventos)
k_values = np.arange(0, 10)  # Rango de valores de k

# Probabilidades para cada valor de k
probabilidades = poisson.pmf(k_values, lambda_param)

# Crear la gr√°fica de barras
plt.plot(k_values, probabilidades)

# Etiquetas y t√≠tulo
plt.xlabel("N√∫mero de eventos (k)")
plt.ylabel("Probabilidad")
plt.title("Distribuci√≥n de Poisson (Œª={})".format(lambda_param))

# Mostrar la gr√°fica
plt.show()
```

La distribuci√≥n uniforme (continua) es la primera distribuci√≥n continua que se estudia por su simpleza. Las distribuciones continuas, a diferencia de las distribuciones de probabilidad discretas, no le asignan probabilidades a puntos de un intervalo, sino m√°s bien a subintervalos como √°reas bajo la gr√°fica de densidad. Vale la pena aclarar que el los t√©rminos densidad y distribuci√≥n son diferentes: el t√©rmino densidad alude a una funci√≥n continua, mientras que la distribuci√≥n de probabilidad asociada es una funci√≥n, cuyo dominio son los subintervalos del dominio de la funci√≥n de densidad. Desde luego, por su definici√≥n misma, las distribuciones de probabilidad continuas se basan en el c√°lculo integral (para poder calcular √°reas bajo la gr√°fica de la funci√≥n de densidad).

```python
# prompt: distribuci√≥n uniforme

import numpy as np
import matplotlib.pyplot as plt

# Distribuci√≥n uniforme
def uniforme(a, b, n):
    """
    Genera una muestra de una distribuci√≥n uniforme.

    Args:
        a: L√≠mite inferior del intervalo.
        b: L√≠mite superior del intervalo.
        n: N√∫mero de muestras a generar.

    Returns:
        Una lista con las muestras generadas.
    """
    return np.random.uniform(a, b, n)

# Par√°metros de la distribuci√≥n uniforme
a = 0  # L√≠mite inferior
b = 10  # L√≠mite superior
n = 1000 # N√∫mero de muestras

# Generar las muestras
muestras = uniforme(a, b, n)


# Crear el histograma
plt.hist(muestras, bins=20, density=True) #bins=20 para tener 20 barras

# Etiquetas y t√≠tulo
plt.xlabel("Valores")
plt.ylabel("Frecuencia")
plt.title(f"Distribuci√≥n Uniforme (a={a}, b={b})")

# Mostrar la gr√°fica
plt.show()
```

```python
# prompt: distribuci√≥n uniforme acumulada

import numpy as np
import matplotlib.pyplot as plt

def uniforme_acumulada(a, b, x):
    """
    Calcula la funci√≥n de distribuci√≥n acumulada (CDF) de una distribuci√≥n uniforme.

    Args:
        a: L√≠mite inferior del intervalo.
        b: L√≠mite superior del intervalo.
        x: Valor para el cual se calcular√° la CDF.

    Returns:
        El valor de la CDF en x.
    """
    if x < a:
        return 0
    elif x >= b:
        return 1
    else:
        return (x - a) / (b - a)

# Ejemplo de uso
a = 0
b = 10
x_values = np.linspace(a - 2, b + 2, 500)  # Valores de x para graficar
cdf_values = [uniforme_acumulada(a, b, x) for x in x_values]

plt.plot(x_values, cdf_values)
plt.xlabel("x")
plt.ylabel("F(x)")
plt.title("Funci√≥n de Distribuci√≥n Acumulada (Uniforme)")
plt.grid(True)
plt.show()
```

La distribuci√≥n normal bivariada es la versi√≥n de la distribuci√≥n normal con dos variables independientes. Su gr√°fico es una campana de Gauss en el espacio tridimensional. Para entenderal en el espacio bidimensional, aprovechamos la funci√≥n contourf, que permite visualizar el gr√°fico a lo largo del eje z desde una altura suficiente para apreciar los diferentes bordes cuyas alturas se identifican con colores distintos.

```python
# prompt: distribuci√≥n normal bivariada

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

# Define the mean vector and covariance matrix
mean = [0, 0]
covariance = [[1, 0.5], [0.5, 1]]

# Create a multivariate normal distribution object
distribution = multivariate_normal(mean=mean, cov=covariance)

# Generate points for the x and y axes
x, y = np.mgrid[-3:3:.01, -3:3:.01]
pos = np.dstack((x, y))

# Calculate the probability density function (PDF) values
pdf = distribution.pdf(pos)

# Plot the PDF as a contour plot
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111)
ax.contourf(x, y, pdf, cmap='viridis')

# Set plot labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Bivariate Normal Distribution')

# Show the plot
plt.show()
```

Esperanza matem√°tica (valor esperado, valor promedio o media aritm√©tica) de una distribuci√≥n de probabilidad discreta.

```python
# prompt: esperanza matem√°tica caso discreto

# Esperanza matem√°tica (valor esperado) para una variable discreta

def esperanza_matematica_discreta(probabilidades, valores):
    """
    Calcula la esperanza matem√°tica (valor esperado) de una variable discreta.

    Args:
        probabilidades: Una lista o array de probabilidades para cada valor de la variable.
        valores: Una lista o array de valores que toma la variable.

    Returns:
        La esperanza matem√°tica (valor esperado) de la variable.
    """

    if len(probabilidades) != len(valores):
        raise ValueError("Las listas de probabilidades y valores deben tener la misma longitud.")

    esperanza = sum(p * x for p, x in zip(probabilidades, valores))
    return esperanza

# Ejemplo de uso:
probabilidades = [0.2, 0.3, 0.1, 0.4]  # Probabilidades
valores = [1, 2, 3, 4]                # Valores correspondientes

esperanza = esperanza_matematica_discreta(probabilidades, valores)
print(f"La esperanza matem√°tica es: {esperanza}")
```

Defina una funci√≥n para calcular el valor esperado de una distribuci√≥n normal.

```python
# prompt: varianza para una distribuci√≥n de probabilidad discreta

import numpy as np

def varianza_discreta(probabilidades, valores):
    """
    Calcula la varianza de una distribuci√≥n de probabilidad discreta.

    Args:
        probabilidades: Una lista o array de probabilidades para cada valor de la variable.
        valores: Una lista o array de valores que toma la variable.

    Returns:
        La varianza de la distribuci√≥n.
    """

    if len(probabilidades) != len(valores):
        raise ValueError("Las listas de probabilidades y valores deben tener la misma longitud.")

    esperanza = sum(p * x for p, x in zip(probabilidades, valores))
    varianza = sum(p * (x - esperanza)**2 for p, x in zip(probabilidades, valores))
    return varianza

# Ejemplo de uso:
probabilidades = [0.2, 0.3, 0.1, 0.4]  # Probabilidades
valores = [1, 2, 3, 4]                # Valores correspondientes

varianza = varianza_discreta(probabilidades, valores)
print(f"La varianza de la distribuci√≥n es: {varianza}")
```

Defina funciones para calcular la varianza de las distribuciones de probabilidad binomial y de Poisson

```python
# prompt: varianza para una distribuci√≥n de probabilidad continua

from scipy import integrate

def varianza_continua(funcion_densidad, esperanza, limite_inferior, limite_superior):
    """
    Calcula la varianza de una distribuci√≥n de probabilidad continua.

    Args:
        funcion_densidad: La funci√≥n de densidad de probabilidad de la variable.
        esperanza: La esperanza matem√°tica de la distribuci√≥n.
        limite_inferior: El l√≠mite inferior del intervalo de integraci√≥n.
        limite_superior: El l√≠mite superior del intervalo de integraci√≥n.

    Returns:
        La varianza de la distribuci√≥n.
    """
    varianza, _ = integrate.quad(lambda x: (x - esperanza)**2 * funcion_densidad(x), limite_inferior, limite_superior)
    return varianza

# Ejemplo de uso con la distribuci√≥n uniforme:

def uniforme_pdf(x, a, b):
  if a <= x <= b:
    return 1 / (b-a)
  else:
    return 0

a = 0
b = 10
esperanza = (a + b) / 2  # Esperanza de la distribuci√≥n uniforme

varianza = varianza_continua(lambda x: uniforme_pdf(x, a, b), esperanza, a, b)
print(f"La varianza de la distribuci√≥n uniforme es: {varianza}")
```

Defina una funci√≥n para calcular la varianza de una distribuci√≥n normal

```python
# prompt: k-√©simo momento para una distribuci√≥n de probabilidad discreta

def k_esimo_momento(probabilidades, valores, k):
    """
    Calcula el k-√©simo momento de una distribuci√≥n de probabilidad discreta.

    Args:
        probabilidades: Una lista o array de probabilidades para cada valor de la variable.
        valores: Una lista o array de valores que toma la variable.
        k: El orden del momento a calcular.

    Returns:
        El k-√©simo momento de la distribuci√≥n.
    """

    if len(probabilidades) != len(valores):
        raise ValueError("Las listas de probabilidades y valores deben tener la misma longitud.")

    momento = sum(p * (x ** k) for p, x in zip(probabilidades, valores))
    return momento

# Ejemplo de uso:
probabilidades = [0.2, 0.3, 0.1, 0.4]  # Probabilidades
valores = [1, 2, 3, 4]                # Valores correspondientes
k = 3  # Orden del momento

momento = k_esimo_momento(probabilidades, valores, k)
print(f"El {k}-√©simo momento de la distribuci√≥n es: {momento}")
```

La covarianza de dos distribuciones de probabilidad discretas es un ejemplo del primer momento producto con respecto a las medias.

```python
# prompt: covarianza de distribuciones de probabilidad discretas

import numpy as np

def covarianza_discreta(probabilidades_conjuntas, valores_x, valores_y):
    """
    Calcula la covarianza de dos variables aleatorias discretas.

    Args:
        probabilidades_conjuntas: Una matriz de probabilidades conjuntas de X y Y.
        valores_x: Una lista o array de valores que toma la variable X.
        valores_y: Una lista o array de valores que toma la variable Y.

    Returns:
        La covarianza de X y Y.
    """

    if not isinstance(probabilidades_conjuntas, np.ndarray):
      probabilidades_conjuntas = np.array(probabilidades_conjuntas)

    esperanza_x = np.sum(probabilidades_conjuntas * valores_x)
    esperanza_y = np.sum(probabilidades_conjuntas * valores_y)
    covarianza = np.sum((valores_x - esperanza_x) * (valores_y - esperanza_y) * probabilidades_conjuntas)
    return covarianza


# Ejemplo de uso
probabilidades = [[0.1, 0.2, 0.1], [0.05, 0.05, 0.2], [0.1, 0.1, 0.1]]
valores_x = [1, 2, 3]
valores_y = [1, 2, 3]
cov = covarianza_discreta(probabilidades, valores_x, valores_y)
print(f"La covarianza es: {cov}")
```
