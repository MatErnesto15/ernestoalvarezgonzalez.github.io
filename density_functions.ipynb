La función gama, $\Gamma: \mathbb{R}^+→\mathbb{R}$, está definida como $Γ(\alpha)=∫_0^{∞}y^{\alpha-1}e^{-y}dy$

Propiedades:

*   $\Gamma(\alpha) = (\alpha-1)\Gamma(\alpha-1)$
*   $\Gamma(1) = 1$
*   $\Gamma(\frac{1}{2})=\sqrt{\pi}$

```python
# prompt: función gamma

import scipy.special

def gamma_function(x):
  """
  Calculates the gamma function for a given value of x.

  Args:
    x: The input value for the gamma function.

  Returns:
    The value of the gamma function at x.
  """
  return scipy.special.gamma(x)
```

```python
gamma_function(1)
```

```python
import numpy as np
gamma_function(1/2)-np.sqrt(np.pi)
```

La distribución gama, que no la función de densidad, tiene dos parámetros, $\alpha>0$ y $\beta>0$, está dada por la fórmula

$
g(x;\alpha,\beta)=
\begin{cases}
\frac{1}{\beta^{\alpha}\Gamma(\alpha)}x^{\alpha-1}e^{-\frac{x}{\beta}} & x>0\\
0 & \text{otro caso}
\end{cases}
$

```python
# prompt: distribución gamma

import numpy as np
import matplotlib.pyplot as plt

def gamma_distribution(x, alpha, beta):
  """
  Calculates the probability density function (PDF) of the gamma distribution.

  Args:
    x: The input value for the PDF.
    alpha: The shape parameter of the gamma distribution.
    beta: The scale parameter of the gamma distribution.

  Returns:
    The value of the PDF at x.
  """
  if x > 0:
    return (1 / (beta**alpha * gamma_function(alpha))) * (x**(alpha - 1)) * np.exp(-x / beta)
  else:
    return 0

# Example usage
alpha = 2
beta = 3
x_values = np.linspace(0, 20, 500)
pdf_values = [gamma_distribution(x, alpha, beta) for x in x_values]

plt.plot(x_values, pdf_values)
plt.xlabel('x')
plt.ylabel('g(x; alpha, beta)')
plt.title(f'Gamma Distribution (alpha={alpha}, beta={beta})')
plt.grid(True)
plt.show()
```

La distribución gama, $g(x;\alpha,\beta)$, permite calcular probabilidades, calculando áreas bajo su gráfica: Por ejemplo, $P(3 \leq x \leq 7) = \int_3^7g(x;\alpha,\beta)dx$.

```python
alpha = 11
beta = 1/5
integral = scipy.integrate.quad(lambda x: gamma_distribution(x, alpha, beta), 3, 7)
print(integral)
```

Como primera salida de resultado, la función scipy.integrate.quad proporciona el área bajo la gráfica sobre el intervalo especificado. La segunda salida de resultado es el error absoluto de estimación de la integral definida.

Distribución exponencial. La distribución exponencial,
$
g(x;\theta) =
\begin{cases}
\frac{1}{\theta}e^{-\frac{x}{\theta}} & x>0 \\
0 & x\leq 0
\end{cases}
$

es una distribución gama con parámetros $\alpha=1$ y $\beta = \theta$.

```python
# prompt: gráfica de la función exponencial

import numpy as np
import matplotlib.pyplot as plt

def exponential_distribution(x, theta):
  """
  Calculates the probability density function (PDF) of the exponential distribution.

  Args:
    x: The input value for the PDF.
    theta: The scale parameter of the exponential distribution.

  Returns:
    The value of the PDF at x.
  """
  if x > 0:
    return (1 / theta) * np.exp(-x / theta)
  else:
    return 0

# Example usage
theta = 2  # Example scale parameter
x_values = np.linspace(0, 10, 500)  # Generate x-values from 0 to 10
pdf_values = [exponential_distribution(x, theta) for x in x_values]

# Plotting
plt.plot(x_values, pdf_values)
plt.xlabel('x')
plt.ylabel('g(x; theta)')
plt.title(f'Exponential Distribution (theta={theta})')
plt.grid(True)
plt.show()
```

La distribución exponencial surge al estimar la probabilidad de observar el primer éxito de un proceso tipo Poisson:

Sea $y>0$ un tiempo fijo (en principio infinitesimalmente pequeño). Suponga que la probabilidad de observar un éxito durante dicho intervalo temporal es $\alpha y$. Sea $p$ la distribución de Poisson con parámetro $\alpha y$. Sea $Y$ la variable aleatoria continua, llamada $\textit{tiempo de espera hasta la ocurrencia del primer éxito}$, con distribución de probabilidad $P$ (y distribución de probabilidad acumulada $F$).

Tenemos lo siguiente:

$
\begin{align}
F(y) = P(Y\leq y) = 1 - P(Y>y) \\
= 1 - P(\text{0 éxitos en el intervalo temporal t}) \\
1 - p(0,;\alpha y) = 1 - \frac{e^{-\alpha y}(\alpha y)^0}{0!} = 1-e^{-\alpha y}.
\end{align}
$


TEOREMA

Si $f(x)$ es una función de densidad continua, y $F(x)$ es la función de distribución acumulada, entonces $P(a\leq X \leq b) = F(b) - F(a)$ y
$f(x) = \frac{dF(x)}{dx}$.

Aplicando este teorema a la función $F(y) = 1-e^{-\alpha y}$, obtenemos que

$f(y) =
\begin{cases}
\alpha e^{-\alpha y} & y > 0 \\
0 & y\leq 0
\end{cases}
$

que es la distribución exponencial con $\theta = \frac{1}{\alpha}$.

La distribución $\chi$-cuadrada, es la distribución gama con $\alpha=\frac{\nu}{2}$ y $\beta = 2$:

$f(x,\nu) =
\begin{cases}
\frac{1}{2^{\frac{\nu}{2}}Γ(\frac{\nu}{2})}x^{\frac{\nu-2}{2}}e^{-\frac{x}{2}} & x>0 \\
0 & x\leq 0
\end{cases}
$

El parámetro $\nu$ se conoce como $\textit{números de grados de libertad}$

El siguiente teorema es importante para derivar fórmulas sobre la media y varianza de la distribución gama, y sobre las distribuciones exponencial y $\chi$-cuadrada:

TEOREMA 2

El $r$-ésimo momento al rededor del origen de la distribución gama está dado por

$\mu'_r = \frac{\beta^rΓ(\alpha + r)}{Γ(\alpha)}$


TEOREMA 3

La media y la varianza de la distribución gama son, respectivamente,

$\mu = \alpha\beta$ y $\sigma^2 = \alpha\beta^2$.
